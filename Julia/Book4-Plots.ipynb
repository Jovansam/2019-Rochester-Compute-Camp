At this point I need to introduce the concept of packages. In Matlab all functionality is built in. Julia is much more modular. There are thousands of packages which do many different things. You can obtain them using the package manager. For example, I installed the PyPlot package which provides plotting functionality, and later we will see packages for optimization and nonlinear equation solving.

To add the GR package from the Julia console, 
1. Hit ] to enter the package manager
2. Type add GR
3. Wait
4. Hit backspace to exit the package manager

To use a package, just type using package.

```julia
using GR
```
